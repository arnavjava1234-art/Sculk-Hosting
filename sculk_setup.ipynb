```python
# Option B: Manual Patch Setup (Use only if HTTP/2 is not default or fails to resolve)
!pip install git+https://github.com/arnavjava1234-art/Sculk-Hosting.git

import sculk_hosting.tunnel
import inspect

file_path = inspect.getfile(sculk_hosting.tunnel)
with open(file_path, "r", encoding="utf-8") as f:
    code = f.read()

target = 'cmd = [self.binary_path, "tunnel", "--url", f"http://127.0.0.1:{self.local_port}"]'
replacement = 'cmd = [self.binary_path, "tunnel", "--protocol", "http2", "--url", f"http://127.0.0.1:{self.local_port}"]'

if target in code:
    code = code.replace(target, replacement)
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(code)
    print("[*] Successfully patched cloudflared to use HTTP/2 protocol!")
else:
    print("[*] cloudflared protocol is already patched or format differs.")

!sculk
```
